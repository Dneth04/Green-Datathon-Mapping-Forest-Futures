# GreenDatathon: Mapping Forest Futures

## Introduction and Event Overview

Welcome to the GreenDatathon: Mapping Forest Futures! This 3-hour data analysis challenge focuses on global forest trends from 1990 to 2020. Our goal is to introduce you to data manipulation and basic programming skills using real-world environmental data.


```python
## Instructions for Loading the Dataset

## To load the dataset, run the following code:

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# If you're using Google Colab, uncomment and run this cell to upload the data
# from google.colab import files
# uploaded = files.upload()

# Load the dataset
df = pd.read_csv('Forest_Area.csv')
```

## Data Exploration
Let's explore the dataset:

```python
# Display the first 5 rows
print(df.head())

# Display basic information about the dataset
print(df.info())

# Display summary statistics
print(df.describe())
```

## Data Filtering
Now, let's filter the data to show countries with large forest areas:

```python
import pandas as pd

# Filter countries with forest area > 1,000,000 ha in 2020
large_forest_countries = df[df['Forest Area, 2020'] > 1000000]

# Print the filtered DataFrame
print(large_forest_countries[['Country and Area', 'Forest Area, 2020']])
```

## Data Grouping and Aggregation
Let's calculate the average forest area:

```python
# Calculate the average forest area for the year 2020
if 'Forest Area, 2020' in df.columns:
    avg_forest_area_2020 = df['Forest Area, 2020'].mean()
    print(f"Average Forest Area in 2020: {avg_forest_area_2020}")
else:
    print("Column 'Forest Area, 2020' is not present in the dataset.")
```

## Data Visualization
Now, let's create a visualization:

```python
# Code for creating a bar plot of the top 10 countries by forest area in 2020
top_10 = df.nlargest(10, 'Forest Area, 2020')
plt.figure(figsize=(12,6))
plt.bar(top_10['Country and Area'], top_10['Forest Area, 2020'])  
plt.title('Top 10 Countries by Forest Area in 2020')
plt.xlabel('Country and Area')
plt.ylabel('Forest Area (1000 ha)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
